## Hyperparameter Tuning Analysis

This notebook analyzes the results of Optuna-based hyperparameter tuning for a PointNet-based 3D bounding box prediction model.



#### Studied Parameters
- Training parameters: batch size, learning rate
- Loss function parameters: weights(center, size, yaw, DIoU)
- validation: val_score = val_loss - 5.0 * val_iou

![Image](/home/void/3d_boundingbox_prediction/outputs/plots/hpo.png)

#### Observations


1. Low learning rates (~1e-4) consistently produce better results

2. Smaller batch sizes (16–32) perform better

3. Loss Weights Have Weak Influence, parameters like `w_size`, `w_yaw`, and `w_center` show no clear pattern

4. IoU around  0.35–0.55, suggests model/loss limitation rather than hyperparameter issue


#### Proposed Improvements

1. Loss Function


2. Data / Sampling : Apply better sampling strategies (FPS)

3. Model Improvements: Upgrade to PointNet++, Add local feature learning
